```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,15,20,25,30]
```

```python
plt.plot(x,y)

plt.title('simple line graph')
plt.xlabel('values of x')
plt.ylabel('values of y')

plt.show()
```

```python
plt.scatter(x,y)

plt.title('scatter plot')
plt.xlabel('values of x')
plt.ylabel('values of y')

plt.show()
```

```python
plt.bar(x,y)

plt.title('bar plot')
plt.xlabel('values of x')
plt.ylabel('values of y')

plt.show()
```

```python
plt.plot(x,y,color='green',linestyle='--',linewidth=2,marker='o',markerfacecolor='red')

plt.title('simple line graph')
plt.xlabel('values of x')
plt.ylabel('values of y')

plt.show()
```
